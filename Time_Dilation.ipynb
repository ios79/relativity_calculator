```python
def calculate_time_dilation(v, t, c=3e8):
    """
    Calculate the dilated time using the Lorentz factor.

    Parameters:
    v (float): Velocity in m/s.
    t (float): Proper time in seconds.
    c (float): Speed of light in m/s (default is 3e8).

    Returns:
    float: Dilated time (t').
    """
    if v >= c:
        return "Velocity cannot exceed or equal the speed of light!"

    gamma = 1 / ((1 - (v**2 / c**2))**0.5)
    t_prime = gamma * t
    return t_prime

# Example usage
v = 290_000 * 1e3  # Velocity in m/s
t = 1  # Proper time in seconds
print(f"Dilated time (t'): {calculate_time_dilation(v, t):.2f} seconds")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def kinetic_energy(v, m, c=3e8):
    """
    Calculate the relativistic kinetic energy.

    Parameters:
    v (float or np.ndarray): Speed in m/s.
    m (float): Mass in kg.
    c (float): Speed of light in m/s (default is 3e8).

    Returns:
    float or np.ndarray: Kinetic energy in Joules.
    """
    return (0.5 * m * v**2) / np.sqrt(1 - (v**2 / c**2))

# Constants
m = 1  # Mass of the object in kg (you can change this value)
c = 3e8  # Speed of light in m/s

# Speed range from 0 to 0.99c
v = np.linspace(0, 0.99 * c, 1000)

# Calculate kinetic energy
K = kinetic_energy(v, m, c)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(v / c, K, label=r"$K = \frac{0.5 m v^2}{\sqrt{1 - \frac{v^2}{c^2}}}$" + "\n" + r"(Relativistic)")
plt.title('Kinetic Energy vs Speed')
plt.xlabel('Speed (v/c)')
plt.ylabel('Kinetic Energy (Joules)')
plt.grid()
plt.legend()
plt.show()
```
